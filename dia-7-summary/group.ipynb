```python
import pandas as pd

transacoes = pd.read_csv("../DB/transacoes.csv", sep=";")

transacoes.head()
```

```python
#Agrupando por cliente e contando as transacoes

# transacoes.groupby(by=["IdCliente"], as_index=False)[["IdTransacao"]].count()

#ou

transacoes.groupby(by=["IdCliente"], as_index=False).agg({
    "IdTransacao":["count"]
})

```

```python
#quantidade de transações, total pontos e pontos/transações por clientes

summary = transacoes.groupby(by=["IdCliente"], as_index=False).agg({

    "IdTransacao":["count"],
    "QtdePontos":["mean", "sum"]
    
})
```
