```python
import pandas as pd

transacoes = pd.read_csv("../DB/transacoes.csv", sep=";")

transacoes.head()
```

```python
import numpy as np

def dif_amp(x: pd.Series):
    amp = x.max()-x.min()
    mean = x.mean()
    return np.sqrt((amp - mean)**2)

def life_time(x: pd.Series):
    df = pd.to_datetime(x)
    return (df.max() - df.min()).days
```

```python
group = transacoes.groupby(by=["IdCliente"]).agg(
    {"IdTransacao":["count"],
     "QtdePontos":["mean", "sum", dif_amp],
     "DtCriacao":[life_time]}

)

group
```
