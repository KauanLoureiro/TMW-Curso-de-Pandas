Selecione a primeira transação diaria de cada cliente

```python
import pandas as pd
```

```python
df = pd.read_csv("../DB/transacoes.csv", sep=";")
```

```python
df.head()
```

```python
df.sort_values(["IdCliente", "DtCriacao"], inplace=True)
df["DtCriacao"] = pd.to_datetime(df["DtCriacao"])
df["dia"]= df["DtCriacao"].dt.date
df.drop_duplicates(subset=["IdCliente", "dia"], inplace=True)
```
