```python
import pandas as pd
df = pd.read_csv("../DB/clientes.csv", sep=';')
```

```python
df.head()
```

```python
#Vamos recortar o idCliente

def get_last_id(x):
    return x.split("-")[-1]
```

```python
get_last_id("002970bd-65c2-42bf-a17d-2bc5e95e2f0b")
```

```python
#Criar uma nova coluna apenas com o ultimo id
id_last_list = []
id_full_list = df["idCliente"].tolist()

for i in id_full_list:
    id_last_list.append(get_last_id(i))

df["id_last_list"] = id_last_list
df.head()
```

```python
#Utilizando o apply

df["id_last_apply"] = df["idCliente"].apply(get_last_id)
df.head()

#Ou seja, o apply aplica uma certa função para cada elemento da coluna. Muito útil.
```
