```python
import requests
import pandas as pd

url = "https://pt.wikipedia.org/wiki/Unidades_federativas_do_Brasil"  

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

dfs = pd.read_html(response.text)

df = dfs[1]

df.head()
```

```python
# Função normalizar

def normalise(x:str):
    x = float(x.replace(" ", "").replace(",", ".").replace("\xa0","").replace("anos",""))
    return x

def normalise_Alfabetização(x:str):
    x = float(x.replace("%","").replace(" ", "").replace(",", "."))
    x = x * 0.01
    return x

def normalise_Mortalidade_infantil(x:str):
    x = float(x.replace(" ", "").replace("‰","").replace(",", "."))
    x = x * 0.001
    return x
```

```python
df["Área (km²)"] = df["Área (km²)"].apply(normalise)
```

```python
df["População (Censo 2022)"] = df["População (Censo 2022)"].apply(normalise)
```

```python
df["PIB (2015)"] = df["PIB (2015)"].apply(normalise)

df["PIB per capita (R$) (2015)"] = df["PIB per capita (R$) (2015)"].apply(normalise)

df["Alfabetização (2016)"] = df["Alfabetização (2016)"].apply(normalise_Alfabetização)

df["Mortalidade infantil (2016)"] = df["Mortalidade infantil (2016)"].apply(normalise_Mortalidade_infantil)

df["Expectativa de vida (2016)"] = df["Expectativa de vida (2016)"].apply(normalise)
df.rename(columns={"Expectativa de vida (2016)": "Expectativa de vida em anos (2016)"}, inplace=True)
```

```python
def regioes(x):
    if x["Unidade federativa"] in ["Distrito Federal","Goiás","Mato Grosso","Mato Grosso do Sul"]:
        return "Centro-Oeste"

    elif x["Unidade federativa"] in ["Alagoas","Bahia","Ceará","Maranhão","Paraíba","Pernambuco","Piauí","Rio Grande do Norte","Sergipe"]:
        return "Nordeste"

    elif x["Unidade federativa"] in ["Acre","Amapá","Amazonas","Pará","Rondônia","Roraima","Tocantins"]:
        return "Norte"

    elif x["Unidade federativa"] in ["Espírito Santo","Minas Gerais","Rio de Janeiro","São Paulo"]:
        return "Sudeste"

    elif x["Unidade federativa"] in ["Paraná","Rio Grande do Sul","Santa Catarina"]:
        return "Sul"

df["Regiões"] = df.apply(regioes, axis=1)
```

```python
def classifica_bom(linha):
    if (linha["PIB per capita (R$) (2015)"] > 30000 and linha["IDH (2010)"] > 700 and linha["Mortalidade infantil (2016)"] < 0.015):
        return "Parece bom"
    else:
        return "Parece ruim"

df["classificacao"] = df.apply(classifica_bom, axis=1)
```
